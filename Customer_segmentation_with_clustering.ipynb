```python
# Calculate age by subtracting the birth date from today's date
today = pd.to_datetime('today')
agg_df['Age'] = (today - agg_df['Customer_Birthdate1']).dt.days // 365


# Calculate Recency
agg_df['Recency'] = (today - agg_df['latest_Order_Date']).dt.days

# Calculate Average Purchase Value (Monetary / Frequency)
agg_df['Average_Purchase_Value'] = agg_df['Total_Spent'] / agg_df['Frequency']

agg_df['Customer_Lifespan'] = agg_df['Recency'] / 365

# Calculate CLV (Average Purchase Value * Frequency * Lifespan)
agg_df['CLV'] = agg_df['Average_Purchase_Value'] * agg_df['Frequency'] * (agg_df['Customer_Lifespan'] / 365)
```

```python
# Import necessary library.
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.cluster import Birch
```

```python
# Elbow method: try different k.
range_n_clusters = [2, 3, 4, 5, 6]

# Select the features for clustering
X = f_df[['Recency',	'CLV',	'Age',	'Frequency',	'avg_unit_cost']]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create an empty list to store the within-cluster sum of squares (WCSS)
wcss=[]

# Iterate over different values of k (number of clusters)
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=0, n_init = 10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

**Silhouette method**


```python
# Range of clusters to evaluate
k_range = range(2, 11)  # Evaluate from 2 to 10 clusters

# Create an empty list to store silhouette scores
silhouette_scores = []

# Iterate through different values of k
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(X_scaled, labels)
    silhouette_scores.append(silhouette_avg)

# Plot the silhouette scores
plt.plot(k_range, silhouette_scores)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Analysis for Optimal K')
plt.show()

# Find the optimal k (index of the highest silhouette score)
optimal_k = k_range[np.argmax(silhouette_scores)]
print(f"Optimal number of clusters (based on silhouette score): {optimal_k}")
```

```python
# Extract feature matrix
X = f_df.values

# Optional: Reduce dimensionality for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Silhouette analysis for various numbers of clusters
range_n_clusters = [2, 3, 4, 5, 6]  # Example range for number of clusters

for n_clusters in range_n_clusters:
    # Create a subplot with 1 row and 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # The 1st subplot is the silhouette plot
    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(X) + (n_clusters + 1) * 10])

    # Initialize the clusterer with n_clusters value and a random generator seed of 10 for reproducibility
    clusterer = KMeans(n_clusters=n_clusters, random_state=10, n_init=10)
    cluster_labels = clusterer.fit_predict(X)

    # The silhouette_score gives the average value for all the samples
    silhouette_avg = silhouette_score(X, cluster_labels)
    print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)

    # Compute the silhouette scores for each sample
    sample_silhouette_values = silhouette_samples(X, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Aggregate the silhouette scores for samples belonging to cluster i and sort them
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label the silhouette plots with their cluster numbers at the middle
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Compute the new y_lower for next plot
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("The silhouette plot for the various clusters.")
    ax1.set_xlabel("The silhouette coefficient values")
    ax1.set_ylabel("Cluster label")

    # The vertical line for average silhouette score of all the values
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])  # Clear the y-axis labels / ticks
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # 2nd Plot showing the actual clusters formed
    colors = cm.nipy_spectral(cluster_labels.astype(float) / n_clusters)
    ax2.scatter(X_pca[:, 0], X_pca[:, 1], marker='.', s=30, lw=0, alpha=0.7,
                c=colors, edgecolor='k')

    # Labeling the clusters
    centers = clusterer.cluster_centers_
    centers_pca = pca.transform(centers)  # Project cluster centers onto PCA space

    # Draw white circles at cluster centers
    ax2.scatter(centers_pca[:, 0], centers_pca[:, 1], marker='o',
                c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers_pca):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1,
                    s=50, edgecolor='k')

    ax2.set_title("The visualization of the clustered data.")
    ax2.set_xlabel("Feature space for the 1st principal component")
    ax2.set_ylabel("Feature space for the 2nd principal component")

    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
                  "with n_clusters = %d" % n_clusters),
                 fontsize=14, fontweight='bold')

    plt.show()
```

**Pipeline for Kmeans**

```python
f_df = pd.DataFrame(f_df)

# Define transformers
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Define numerical features
numeric_features = ['Recency', 'CLV', 'Age', 'Frequency', 'avg_unit_cost']

# Create ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features)
    ]
)

# Create K-Means clustering pipeline
kmeans_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('kmeans', KMeans(n_clusters=5, random_state=42, n_init=10))
])

# Fit and predict using K-Means
f_df['kmeans_labels'] = kmeans_pipeline.fit_predict(f_df)
```

**Pipeline for Hierarchical**

```python
f_df = pd.DataFrame(f_df)

# sampled dataset for Hierarchical clustering

sampled_df = f_df.sample(n=5000, random_state=42)

sampled_df = pd.DataFrame(sampled_df)
sampled_df = sampled_df.drop('kmeans_labels', axis=1)

# Define transformers
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Define numerical features
numeric_features = ['Recency', 'CLV', 'Age', 'Frequency', 'avg_unit_cost']

# Create ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features)
    ]
)

# Create Hierarchical clustering pipeline
hierarchical_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('hierarchical', AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='average'))
])

# Fit and extract labels using Hierarchical Clustering
hierarchical_pipeline.fit(sampled_df)
sampled_df['hierarchical_labels'] = hierarchical_pipeline.named_steps['hierarchical'].labels_
```
